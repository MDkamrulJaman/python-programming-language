##Labels and Title for a Plot

*   markersize      - ms
*   markeredgecolor - mec
*   markerfacecolor - mfc

```python
x_axis = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y_axis = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x_axis, y_axis, color = 'g', linewidth = '3', marker='o',  ms=20, mfc="b", mec="r" )

# Title
plt.title("Data of Velocity")

#x_axis label name
plt.xlabel("Velocity")

#y_axis label name
plt.ylabel("Time")

#show the graph
plt.show()
```

# Grid inside graph


* using fontdict




```python
x_axis = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y_axis = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])



font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}



# Plot the graph
plt.plot(x_axis, y_axis, color = 'g', linewidth = '3', marker='o',  ms=20, mfc="b", mec="r" )

# Title
plt.title("Data of Velocity", fontdict = font1)

#x_axis label name
plt.xlabel("Velocity - horizontal", fontdict = font2)

#y_axis label name
plt.ylabel("Time - vertical ", fontdict = font2)

# Grid
plt.grid(color = 'g', linestyle = 'dashed', linewidth = 1)

#show the graph
plt.show()
```

#Subplot() display Multiple Plots


*   subplot(row, coloum, index of the current plot)
*   subplot(row = 1, coloum = 2, index of the current plot = 1)



Display coloum-wise

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# plot_1
plt.subplot(1, 2, 1)
plt.plot(x,y)


x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# plot_2
plt.subplot(1,2,2)
plt.plot(x,y)



plt.show()


```

Display row-wise

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# plot_1
plt.subplot(2, 1, 1)
plt.plot(x,y)


x = np.array([0, 1, 2, 3])
y = np.array([20, 30, 40, 500])

# plot_2
plt.subplot(2,1,2)
plt.plot(x,y)



plt.show()

```

Display multiple figure

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for plots
data = [
    (np.array([0, 1, 2, 3]), np.array([3, 8, 1, 10])),
    (np.array([0, 1, 2, 3]), np.array([10, 20, 30, 40])),
    (np.array([0, 1, 2, 3]), np.array([3, 8, 1, 10])),
    (np.array([0, 1, 2, 3]), np.array([10, 20, 30, 40])),
    (np.array([0, 1, 2, 3]), np.array([3, 8, 1, 10])),
    (np.array([0, 1, 2, 3]), np.array([10, 20, 30, 40]))
]

# subplots using loop
for i, (x, y) in enumerate(data, start=1):
    plt.subplot(2, 3, i)
    plt.plot(x, y)

plt.tight_layout() # for responsive layout
plt.show()
```

#  Bar chart

```python
x = np.array(['A','B','C','D','E','F','G','H','I'])

y = np.array([10,20,30,40,50,60,70,80,90])


# Vertical View
plt.subplot(1,2,1)
plt.bar(x, y , color ='green', width = 0.2) # width for bar
plt.title('Vertical')
plt.xlabel('Name')
plt.ylabel('Amplitude')


# Horizontal View
plt.subplot(1,2,2)
plt.barh(x, y, color ='red', height = 0.1) # height for barh
plt.title('Horizontal')
plt.xlabel('Name')
plt.ylabel('Amplitude')

plt.grid()
plt.tight_layout()
plt.show()
```

# Pie chart

```python
x = np.array([10,20,30,40])

plt.pie(x)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])

mylabels = ["M1", "K2", "J3", "R4"]

mycolors = ["red", "hotpink", "b", "#4CAF50"]

myexplode = [0.1, 0.2, 0.3, 0.4]

plt.pie(y, labels = mylabels, explode = myexplode, colors= mycolors, shadow = True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'Apple', 'Banana', 'Cherry', 'Date'

sizes = [215, 130, 245, 210]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

explode = (0.1, 0, 0, 0)  # explode the 1st slice

plt.figure(figsize=(6,6)) #(horizontal, vertical)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)

plt.axis('equal')  # For equal aspect ratio
plt.title('Fruit Pie Chart')

plt.show()
```
